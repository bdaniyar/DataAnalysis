```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('StudentsPerformance.csv')

print("First 5 rows of the dataset:")
display(df.head()) 

print("\nDataset Information:")
df.info()
```

```python
correlations = df.corr(numeric_only=True)

print("Correlation Matrix:")
display(correlations)

plt.figure(figsize=(8, 6))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Test Scores")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='reading score', y='writing score', hue='gender')
plt.title('Relationship between Reading and Writing Scores')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.grid(True)
plt.show()
```

The analysis of the student performance dataset reveals a very strong positive correlation between reading and writing scores (approximately 0.95). This suggests that these two literacy skills are deeply interconnected; students who excel in reading are almost certain to perform well in writing. While math scores also show a high correlation with reading (0.82), the link is slightly weaker than the one between the two verbal subjects. The visualization confirms a clear linear trend, indicating that educational interventions aimed at improving reading comprehension will likely have a direct positive impact on writing proficiency as well.
